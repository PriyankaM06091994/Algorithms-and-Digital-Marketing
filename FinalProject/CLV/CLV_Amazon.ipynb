```
df1['user_Id'] = df1.groupby(['reviews.username'], sort=False).ngroup().add(1)
```

```
df1['Total_Visits'] = np.random.randint(1, 50, df1.shape[0])
```

```
df1[['reviews.username','user_Id','Total_Visits']].shape
```

```
df2=df1[['reviews.username','user_Id','Total_Visits']]
```

```
df2
```

```
import random
```

```
df2['Total_Price'] = np.random.randint(100,1200,size=34660)/10
```

```
df2['Total Profit'] = np.random.randint(2,9,size=34660)/10
```

```
df2.rename(columns = {'reviews.username': 'User Name'}, inplace=True)
```

```
df2
```

```
df2['Customer Lifetime value']=(df2['Total_Price']*df2['Total Profit']*df2['Total_Visits']).astype(int)
```

```
df2
```

```
import matplotlib.pyplot as plt
```

```
df2['Customer Lifetime value'].plot(kind='hist')
```

```
Membership = [df2['Customer Lifetime value'].between(0, 500), 
            df2['Customer Lifetime value'].between(500, 1500), 
            df2['Customer Lifetime value'].between(1500, 3000),
            df2['Customer Lifetime value'].between(3000, 10000)]
values = ['Regular', 'Prime', 'Prime Gold', 'Prime Platinum']

df2['CLV_Category'] = np.select(Membership, values, 'Prime Gold' )
```

```
df2.head(100)
```

```
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
```

```
plt.figure(figsize=(8,4))
sns.countplot(x="CLV_Category", data=df2, color=color[0])
plt.ylabel('Count', fontsize=12)
plt.xlabel('Categories', fontsize=12)
plt.xticks(rotation='vertical')
plt.title("Users In each Category", fontsize=15)
plt.show()
```
